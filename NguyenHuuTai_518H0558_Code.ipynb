```python
#1.2 a
result = []
for i in range (1,5):
  resp = requests.get("http://quotes.toscrape.com/page/" + str(i) +"/")
  soup = BeautifulSoup(resp.content, "html.parser")
  resu = soup.findAll('div',attrs = {'class':'quote'})
  result.append(resu)
for res in result:
  for re in res:
    print(re.text)
```

```python
#1.2 c
authors = []
for resul in result:
  for res in resul:
    author = {}
    author['Tacgia'] = res.find('small', attrs={'class':'author'}).text
    link = "https://quotes.toscrape.com/"+res.a['href']
    author['Link'] = link
    response1 = requests.get(link)
    soup1 = BeautifulSoup(response1.content, "html.parser")
    borndate = soup1.find('span', attrs={'class':'author-born-date'})
    author['Namsinh'] = borndate.text
    author['Quote'] = res.find('span', attrs={'class':'text'}).text
    authors.append(author)

print(authors)
```

```python
import string
from nltk.corpus import stopwords

def text_process(mess):
    STOPWORDS = stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure']
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]
    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    # Now just remove any stopwords
    return ' '.join([word for word in nopunc.split() if word.lower() not in STOPWORDS])
```

```python
threshold = 0.1
```

```python
for x in range(0,X.shape[0]):
  for y in range(x,X.shape[0]):
    if(x!=y):
      if(cosine_similarity(X[x],X[y])>threshold):
        print(dsf["Tacgia"][x],":",corpus[x])
        print(dsf["Tacgia"][y],":",corpus[y])
        print("Cosine similarity:",cosine_similarity(X[x],X[y]))
        print()
```
